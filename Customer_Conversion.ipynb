### Description of columns:

**PageVisitsCatX:** Number of pages visited by a visitor. Cat1 is administrative, Cat2 is informational, and Cat3 is product related.

**BounceRates:** Percentage of visitors who landed and exited a page.

**PageValues:** A measurement of a page's contribution to a sale.

**HowCloseToSpecialDay:** How close is browsing date to a special day or a holiday, such as Valentine's day? Higher numbers are closer.

**Month:** Month of visit.

**VisitorType:** Is the visitor a new or returning visitor?

**Weekend:** Did the visit occur on a weekend day?

**Revenue:** Did the visit conclude in a sale?

## Perform Logistic Regression 

```python
df1 = df.drop(['Month','VisitorType','Revenue'],axis=1)
df1.head()
```

```python
# Convert categorical variables to binary variables 
Visitor_dummies = pd.get_dummies(df['VisitorType'])

# Review the DataFrame
Visitor_dummies.tail()
```

```python
month_dummies = pd.get_dummies(df['Month'])

# Review the DataFrame
month_dummies.tail()
```

```python
df2 = pd.concat([df1,Visitor_dummies,month_dummies],axis=1)
df2.head()
```

```python
# List the columns
df.columns
```

```python
# Separate the features and target variables.
X = df2.copy()
y = df['Revenue']
```

```python
# Split the dataset into training and testing sets
X_train,X_test,y_train,y_test = train_test_split(X,y, random_state=42)
```

```python
# Instantiate a logistic regression model
model = LogisticRegression(solver='lbfgs',random_state = 42)
```

```python
# Train the model
model.fit(X_train,y_train)
```

```python
# Use the testing data to make predictions.
log_prediction = model.predict(X_test)
result = pd.DataFrame({"Predictions":log_prediction,"Actual":y_test}).reset_index(drop=True)
result.head()
```

```python
# Calculate the accuracy. 
print(f" Training_score: {model.score(X_train,y_train)}")
print(f" Test_score: {model.score(X_test,y_test)}")
```

```python
# Print the confusion matrix.
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test, log_prediction)
```

```python
# Print a classification report
from sklearn.metrics import classification_report
print(classification_report(y_test, log_prediction))
```
